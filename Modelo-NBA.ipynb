```python
#dataframe com as informações dos players
colunas = list(df_all_temps.columns.tolist())
df_players = pd.DataFrame(columns = colunas)

#lista de players
lista_players = list(df_all_temps['Player'].unique())

#inserindo a linha de cada player
for player in lista_players:
  info_p = []
  for coluna in colunas:
    if coluna == 'Player':
      info_p.append(player)

    elif coluna == 'Team':
      info_p.append(list(set(list(df_all_temps[df_all_temps['Player'] == player][coluna].tolist()))))

    elif coluna not in ('FG%','3P%', 'FT%','temporada','Player','Team','Age'):
      info_p.append(df_all_temps[df_all_temps['Player'] == player][coluna].sum())

    else:
      info_p.append(df_all_temps[df_all_temps['Player'] == player][coluna].tolist())
  df_players.loc[len(df_players)] = info_p
df_players
```

```python
df_intermed = df_players
```

```python
percent_win = []
percent_loss = []
percent_min = []
percent_pts  = []
percent_FG = []
percent_3P = []
percent_FT = []
percent_PlsMinus = []
percent_OREB = []
percent_DREB = []
percent_REB = []
percent_AST = []
percent_TOV = []
percent_STL = []
percent_BLK = []
percent_PF = []
percent_FP = []

for player in lista_players:
  win = df_intermed[df_intermed['Player']== player]['W'].iloc[0]
  lost = df_intermed[df_intermed['Player']== player]['L'].iloc[0]
  points = df_intermed[df_intermed['Player']== player]['PTS'].iloc[0]
  min = df_intermed[df_intermed['Player']== player]['Min'].iloc[0]
  FMG = df_intermed[df_intermed['Player']== player]['FGM'].iloc[0]
  FMA = df_intermed[df_intermed['Player']== player]['FGA'].iloc[0]
  PM_3 = df_intermed[df_intermed['Player']== player]['3PM'].iloc[0]
  PA_3 = df_intermed[df_intermed['Player']== player]['3PA'].iloc[0]
  FTM = df_intermed[df_intermed['Player']== player]['FTM'].iloc[0]
  FTA = df_intermed[df_intermed['Player']== player]['FTA'].iloc[0]
  PlsMinus = df_intermed[df_intermed['Player']== player]['+/-'].iloc[0]
  OREB = df_intermed[df_intermed['Player']== player]['OREB'].iloc[0]
  DREB = df_intermed[df_intermed['Player']== player]['DREB'].iloc[0]
  REB = df_intermed[df_intermed['Player']== player]['REB'].iloc[0]
  AST = df_intermed[df_intermed['Player']== player]['AST'].iloc[0]
  TOV = df_intermed[df_intermed['Player']== player]['TOV'].iloc[0]
  STL = df_intermed[df_intermed['Player']== player]['STL'].iloc[0]
  BLK = df_intermed[df_intermed['Player']== player]['BLK'].iloc[0]
  PF = df_intermed[df_intermed['Player']== player]['PF'].iloc[0]
  FP = df_intermed[df_intermed['Player']== player]['FP'].iloc[0]
  DD2 = df_intermed[df_intermed['Player']== player]['DD2'].iloc[0]
  TD3 = df_intermed[df_intermed['Player']== player]['TD3'].iloc[0]
  gp = df_intermed[df_intermed['Player']== player]['GP'].iloc[0]

  percent_win.append(win / gp)
  percent_loss.append(lost / gp)
  percent_pts.append(points / gp )
  percent_min.append(min/ gp)
  percent_FG.append(FMG /FMA)
  percent_3P.append(PM_3/PA_3)
  percent_FT.append(FTM/FTA)
  percent_PlsMinus.append(PlsMinus/gp)
  percent_OREB.append(OREB/ gp)
  percent_DREB.append(DREB/ gp)
  percent_REB.append(REB/ gp)
  percent_AST.append(AST/ gp)
  percent_TOV.append(TOV/ gp)
  percent_STL.append(STL/ gp)
  percent_BLK.append(BLK/ gp)
  percent_PF.append(PF/ gp)
  percent_FP.append(FP/ gp)
```
